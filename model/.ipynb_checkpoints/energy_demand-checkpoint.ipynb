```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("data/energy.csv", low_memory=False)
df
```

```python
cols = [1, 3, 5, 7, 9, 11]
titles = [
    "Michigan Coal Consumption",
    "Michigan Natural Gas Consumption",
    "Michigan Nuclear Consumption",
    "Michigan Hydro Consumption",
    "Michigan Wind Consumption",
    "Michigan Solar Consumption"
]

fig, axes = plt.subplots(2, 3, figsize=(15, 8))
axes = axes.flatten()

for ax, i, title in zip(axes, cols, titles):
    x = np.arange(df.iloc[6:, i].size)
    y = np.array(df.iloc[6:, i]).astype(float)
    y = y[::-1]   # reverse order if dataset is descending
    
    ax.plot(x, y, linewidth=1.5)
    ax.set_title(title, fontsize=12, weight="bold")
    ax.set_ylabel("Thousand MWh")
    ax.set_xlabel("Monthly: 1/2001 to 06/2025")

plt.tight_layout()
plt.show()
```
