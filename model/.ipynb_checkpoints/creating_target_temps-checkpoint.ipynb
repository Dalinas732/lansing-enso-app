```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from lansing_temps import Temperature_Data


df_local_temps = pd.read_excel("data/cleandata.xlsx",sheet_name=None)
tp = Temperature_Data(df_local_temps)
years = tp.years
years = years[::-1]
```

```python
### Stores the averaged temperature for a given month for a given year in matrix
highs = np.zeros((12,len(years)))
avgs = np.zeros((12,len(years)))
lows = np.zeros((12,len(years)))
months = np.arange(1,13)
for month in months:
    for i, year in enumerate(years):
        temp = np.mean(tp.collect_month(month,year,0))
        highs[month-1,i] = temp
        temp = np.mean(tp.collect_month(month,year,1))
        avgs[month-1,i] = temp
        temp = np.mean(tp.collect_month(month,year,2))
        lows[month-1,i] = temp
print("Completed.")
```

```python
# Making high columns
highs = np.zeros(12*len(years))
months = np.arange(1,13)

i = 0
for year in years:
    for month in months:
        temp = np.mean(tp.collect_month(month,year,0))
        highs[i] = temp
        i = i + 1 if i != highs.size - 1 else i
```

```python
# Making lows column
lows = np.zeros(12*len(years))
months = np.arange(1,13)

i = 0
for year in years:
    for month in months:
        temp = np.mean(tp.collect_month(month,year,2))
        lows[i] = temp
        i = i + 1 if i != lows.size - 1 else i
```

```python
df_high = pd.DataFrame(highs, columns=["high"])
df_high.to_csv("high.csv", index=False)
```

```python
df_low = pd.DataFrame(lows, columns=["low"])
df_low.to_csv("low.csv", index=False)
```
