```python
# Read CSV, skip metadata rows
df = pd.read_csv("data/Retail_sales_of_electricity_monthly.csv", skiprows=4)
df
```

```python
y = np.array(df.iloc[:,1]).astype(float)
y = y[::-1]
x = np.arange(y.size)

plt.figure(figsize=(12,6))
plt.plot(x, y, color="royalblue", linewidth=2)

# Titles and labels
plt.title("Michigan Retail Sales of Electricity (All Sectors)", fontsize=16, weight="bold")
plt.ylabel("Million kilowatt-hours", fontsize=12)
plt.xlabel("Jan 2001 – Jun 2025", fontsize=12)

# Add gridlines for readability
plt.grid(True, linestyle="--", alpha=0.6)

# Tweak axes
plt.tight_layout()
plt.show()
```
